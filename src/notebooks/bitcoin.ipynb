```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
# Lista de criptomonedas a consultar
tickers = ["BTC-USD","ETH-USD","USDT-USD","XRP-USD","BNB-USD","SOL-USD","DOGE-USD","ADA-USD","TRX-USD","LTC-USD"]
#tickers = ["ETH-USD","USDT-USD","XRP-USD","BNB-USD","SOL-USD","DOGE-USD","ADA-USD","TRX-USD","LTC-USD"]
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=365)).strftime('%Y-%m-%d')
```

```python
data = yf.download(tickers, start=start_date, end=end_date)
```

```python
data_price = data["Close"]
data_price.head(4)
```

```python

plt.figure(figsize=(16, 8))
for ticker in tickers:
    plt.plot(data_price.index, data_price[ticker], label=ticker)

plt.title('Evolucin del Precio de Cierre de Acciones (ltimos 12 meses)')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.legend(loc='upper left', fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.show()
```
